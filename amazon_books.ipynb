### Main question: does the length of the book determine how successful it will be?

```python
import pandas as pd
df = pd.read_csv('final_book_dataset_kaggle2.csv')
df
```

### the column for pages has phrases in some of the rows. Need to be cleaned to convert data to numeric value

```python
# Convert 'pages' column to numeric, coercing errors to NaN to clean out rows with mistakes
df['pages'] = pd.to_numeric(df['pages'], errors='coerce')
df['pages']
```

```python
df.dropna(subset=['avg_reviews',  'pages'], inplace=True)
df #no missing value for our columns of interest.
```

### The '%' sign at the end of the star rating causes the columns to be recognized as a string type. Removing sign and changing to numeric:

```python
df['star5'] = pd.to_numeric(df['star5'].str.replace('%', ''), errors='coerce')
df['star4'] = pd.to_numeric(df['star4'].str.replace('%', ''), errors='coerce')
df['star3'] = pd.to_numeric(df['star3'].str.replace('%', ''), errors='coerce')
df['star2'] = pd.to_numeric(df['star2'].str.replace('%', ''), errors='coerce')
df['star1'] = pd.to_numeric(df['star1'].str.replace('%', ''), errors='coerce')
```

### The data we’ve seen doesn’t support the claim that people give significantly higher reviews or stars to shorter books. That is the ultimate conclusion of our ad-hoc analysis.
